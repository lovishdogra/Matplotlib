```python
import matplotlib.pyplot as plt

year = [1950,1970,1993,2100]
pop = [2.538,3.564,4.567,10.85]

plt.plot(year,pop)

# Defining axis labels
plt.xlabel('Year')
plt.ylabel('Population')

# Graph Title
plt.title('World Population projection')

plt.show()
```
